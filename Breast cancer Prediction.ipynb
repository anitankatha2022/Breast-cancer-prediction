# Training and Testing data

```python
x = dataset.drop(columns = 'diagnosis')
y = dataset['diagnosis']
```

```python
#train_test_splitting of the dataset
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(len(x_train))
print(len(x_test))
```

# Models

# DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
model_1 = DecisionTreeClassifier(max_depth=6, random_state=78)

model_1.fit(x_train,y_train)
```

```python
y_pred=model_1.predict(x_test)
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,mean_squared_error
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print("Training Score: ",model_1.score(x_train,y_train)*100)
```

```python
print(accuracy_score(y_test,y_pred)*100)
```

# LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
model_2= LogisticRegression()
model_2.fit(x_train,y_train)  
```

```python
y_pred=model_2.predict(x_test)
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,r2_score
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print("Training Score: ",model_2.score(x_train,y_train)*100)
```

```python
print(accuracy_score(y_test,y_pred)*100)
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
model_3 = GaussianNB()
model_3.fit(x_train,y_train)
```

```python
y_pred=model_3.predict(x_test)
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,mean_squared_error,r2_score
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print("Training Score: ",model_3.score(x_train,y_train)*100)
```

```python
print(model_3.score(x_test,y_test))
```
